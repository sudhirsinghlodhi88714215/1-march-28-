# Example

```python
import seaborn as sns
```

```python
sns.load_dataset('iris')
```

```python
df=sns.load_dataset('iris')
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
minmax.transform(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
da=minmax.transform(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
import pandas as pd
pd.DataFrame(da,columns=['a','b','c','d'])
```
